# Data Visualization


## MatPlotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
```

```python

y
```

```python
#functional way of creating plot
plt.plot(x,y,'r-')
# plt.show() - in jupiter not needed
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('square fn plot')
```

```python
#mulitple plots
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method way - 2nd way
fig = plt.figure() #like a canvas
#[l b w h] in % range 0 to 1
axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,label='x-y')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('sq fn plot')

axes.plot(y,x,label='y-x')
axes.legend(loc=1) #loc =1,2,3,4
```

```python
#multiple plots
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y,'r--',label='x-y')
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes1.set_title('sq fn plot')

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x,'g*-',label='y-x')
axes2.set_xlabel('x axis')
axes2.set_ylabel('y axis')
axes2.set_title('sq fn plot')
```

```python
#subplots

fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(y,x)
axes[0].set_title('sq fn')
axes[1].plot(y,x)
plt.tight_layout()

#for cur_ax in axes:
#    cur_ax.plot(x,y)
```


### figure size and DPI

```python
fig = plt.figure(figsize=(2,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(4,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
```
