# Seaborn

### statistical plotting library

## 1.Distribution Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
#kind = hex,reg,kde
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
#for all pairas in dataset
```

```python
sns.rugplot(tips['total_bill'])

#rug is dashes...more dashes more density
```

## 2.Categorical plots

```python
sns.barplot(x='sex',y='total_bill',data=tips)
#sex data is categorical ,male . of emale
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
#combination of strip and violin
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

## 3.Matrix Plots 

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tips.head()
```

```python
#for heat map colm and index show indicate a relavent value
```

```python
tc=tips.corr()
tc
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fp=flights.pivot(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidth=1)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

## 4.Grids

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
import matplotlib.pyplot as plt
g=sns.pairplot(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g=sns.factorplot(data=tips,col='time',row='smoker')
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```
